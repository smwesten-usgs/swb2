```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

```

```python
act_et_ncfile = r'test_data/tcaap/tcaap__actual_et__2015-01-01_to_2019-12-31__260_by_180.nc'
act_et_jun = r'test_data/tcaap/monthly_mean_over_all_Jun__actual_et__2015-01-01_to_2019-12-31__180_by_260__inches.asc'
```

```python
act_et_swb_jun = np.loadtxt(act_et_jun, skiprows=6)
```

```python
act_et = xr.open_dataset(act_et_ncfile)
act_et_monthly = act_et['actual_et'].resample(time="M").sum(dim="time")
act_et_monthly_mean = act_et_monthly.groupby("time.month").mean(dim="time")
plt.imshow(act_et_monthly_mean[5])
plt.colorbar()
```

```python
plt.imshow(act_et_swb_jun-act_et_monthly_mean[5])
plt.colorbar()
```
